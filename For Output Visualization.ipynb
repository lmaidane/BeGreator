```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import re
from nltk.util import ngrams
import os
import docx2txt
from geopy.geocoders import Nominatim
from geotext import GeoText
from geopy.distance import great_circle
import csv
```

```python
#Reading languages csv file
data=pd.read_csv("Output3.csv")

#Collecting skills and names and converting to list
skill_list=data['Skills']
names_list=data['Name']
skill_list=list(skill_list)
names_list=list(names_list)
#print(skill_list)
#print(names_list)

li=[]

for i in range(len(skill_list)):
    skill_list[i]=skill_list[i].replace("['","")
    skill_list[i]=skill_list[i].replace("']","")
    li.append(list(skill_list[i].split("', '")))
    
for i in range(len(li)):
    for j in range(len(li[i])):
        print(li[i][j])
    print("\n")
```

```python
r=[]
python=[]
statistics=[]
machine_learning=[]
data_analytics=[]

r_names=[]
python_names=[]
statistics_names=[]
machine_learning_names=[]
data_analytics_names=[]

for i in range(len(li)):
    for j in range(len(li[i])):
        if(li[i][j]=="r"):
            r_names.append(names_list[i])
            r.append("r")
        if(li[i][j]=="python"):
            python_names.append(names_list[i])  
            python.append("python")
        if(li[i][j]=="statistics"):
            statistics_names.append(names_list[i])
            statistics.append("statistics")
        if(li[i][j]=="machine learning"):
            machine_learning_names.append(names_list[i])
            machine_learning.append("machine learning")
        if(li[i][j]=="data analytics"):
            data_analytics_names.append(names_list[i])
            data_analytics.append("data analytics")
            
print(r_names) 
print(python_names)
print(statistics_names)
print(machine_learning_names)
print(data_analytics_names)
```

```python
r_python=[]
stat_ml=[]
r_stat=[]
r_ml=[]
python_stat=[]
python_ml=[]
python_stat_ml=[]
r_stat_ml=[]

r_python_names=[value for value in r_names if value in python_names]
print(r_python_names)
for i in range(len(r_python_names)):
    r_python.append("r & python")

stat_ml_names=[value for value in statistics_names if value in machine_learning_names]
print(stat_ml_names)
for i in range(len(stat_ml_names)):
    stat_ml.append("statistics & machine learning")

r_stat_names=[value for value in r_names if value in statistics_names]
print(r_stat_names)
for i in range(len(r_stat_names)):
    r_stat.append("r & statistics")

r_ml_names=[value for value in r_names if value in machine_learning_names]
print(r_ml_names)
for i in range(len(r_ml_names)):
    r_ml.append("r & machine learning")

python_stat_names=[value for value in statistics_names if value in python_names]
print(python_stat_names)
for i in range(len(python_stat_names)):
    python_stat.append("python & statistics")

python_ml_names=[value for value in python_names if value in machine_learning_names]
print(python_ml_names)
for i in range(len(python_ml_names)):
    python_ml.append("python & machine learning")

python_stat_ml_names=[value for value in python_names if value in stat_ml_names]
print(python_stat_ml_names)
for i in range(len(python_stat_ml_names)):
    python_stat_ml.append("python, statistics & machine learning")

r_stat_ml_names=[value for value in r_names if value in stat_ml_names]
print(r_stat_ml_names)
for i in range(len(r_stat_ml_names)):
    r_stat_ml.append("r, statistics & machine learning")
```
